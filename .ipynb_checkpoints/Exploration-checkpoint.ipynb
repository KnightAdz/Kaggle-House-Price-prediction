## Skewed features?
Do we have any features that are significantly skewed, which may benefit from a logarithmic transform to make them more normally distributed? This should aid the accuracy of a linear regression model. Let's start by calculating skews of the original data

```python
df.skew().sort_values(ascending=False)
```

MiscVal seems to have significant skew, and is defined as "$Value of miscellaneous feature". Let's plot it to take a look

```python
sns.distplot(df['MiscVal'], hist=False);
```

This looks messy, let's try a less extreme variable to start with

```python
sns.distplot(df['LotFrontage'], hist=False);
```

```python
sns.distplot(np.log1p(df['LotFrontage']), hist=False);
```

```python
skews = df.skew()
```

```python
skews = skews[skews > 1]
```

```python
skews = skews.index
```

```python
df[skews].shape[1]
```

```python
df[skews] = np.log1p(df[skews])
```
